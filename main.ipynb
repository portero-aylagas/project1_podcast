```python
import gradio as gr

def run(source_type, text_input, pdf_file, url_input, style, audience):
    # Route to correct source based on selection
    if source_type == "Text":
        content = load_source("text", content=text_input)
    elif source_type == "PDF":
        content = load_source("pdf", file_path=pdf_file.name if pdf_file else None)
    elif source_type == "URL":
        content = load_source("url", url=url_input)          # ← URL goes here

    script     = generate_script(content, style, audience)
    audio      = generate_audio(script, "outputs/test.mp3")
    preview    = "\n\n".join(
        f"**{l['speaker']}:** {l['text']}" for l in script["lines"]
    )
    return audio, preview

demo = gr.Interface(
    fn=run,
    inputs=[
        gr.Radio(["Text", "PDF", "URL"], value="Text", label="Source Type"),
        gr.Textbox(label="Paste text here", lines=5),        # ← Text input
        gr.File(label="Upload PDF", file_types=[".pdf"]),    # ← PDF input
        gr.Textbox(label="Paste URL here",                   # ← URL input
                   placeholder="https://..."),
        gr.Dropdown(["educational", "news", "interview"],
                    label="Style", value="educational"),
        gr.Textbox(label="Target Audience",
                   value="general public"),
    ],
    outputs=[
        gr.Audio(label="Generated Podcast"),
        gr.Markdown(label="Script Preview"),
    ],
    title="🎙️ AI Podcast Studio",
    description="Paste text, upload a PDF, or drop a URL — get a podcast in seconds.",
)

demo.launch()
```
